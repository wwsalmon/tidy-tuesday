```python
import pandas as pd

df = pd.read_csv("./babynames.csv")

df.head()
```

```python
df.describe()
```

```python
df_female = df[df.sex=="F"]
df_male = df[df.sex=="M"]
```

```python
df_female.to_json("./female.json", orient="records")
df_male.to_json("./male.json", orient="records")
```

# 1. Popular names over time

Let's get the 20 most popular male and female names of all time, then graph their popularity

```python
name_counts_male = df_male[["name", "n"]].groupby(["name"]).sum().sort_values("n", ascending=False)
name_counts_male
```

```python
top_male_names = name_counts_male.head(20).reset_index()["name"].to_numpy()
top_male_names
```

```python
df_male_top = df_male.query("name in @top_male_names")
df_male_top
```

```python
df_male_top.to_json("./male-top.json", orient="records")
```

```python
name_counts_female = df_female[["name", "n"]].groupby(["name"]).sum().sort_values("n", ascending=False)
name_counts_female
```

```python
top_female_names = name_counts_female.head(20).reset_index()["name"].to_numpy()
top_female_names
```

```python
df_female_top = df_female.query("name in @top_female_names")
df_female_top.to_json("./female-top.json", orient="records")
```
