```python
import pandas as pd

df = pd.read_csv("./sports.csv", low_memory=False)
```

```python
df.describe()
```

```python
# get just the columns we care about
df_filtered = df[["year", "unitid", "sportscode", "sum_partic_men", "sum_partic_women", "exp_men", "exp_women", "total_exp_menwomen"]].copy()
df_filtered.describe()
```

```python
# find entries that are present in both years' datasets
# do this by comparing unitid and sportscode

df_2015 = df_filtered[df_filtered["year"] == 2015]
unitids_2015 = list(df_2015["unitid"])
sportscode_2015 = list(df_2015["sportscode"])

df_2019 = df_filtered[df_filtered["year"] == 2019]
unitids_2019 = list(df_2019["unitid"])
sportscode_2019 = list(df_2019["sportscode"])

df_filtered = df_filtered[
    ((df_filtered["year"] == 2015) & (df_filtered["unitid"].isin(unitids_2019)) & (df_filtered["sportscode"].isin(sportscode_2019)))
     |
    ((df_filtered["year"] == 2019) & (df_filtered["unitid"].isin(unitids_2015)) & (df_filtered["sportscode"].isin(sportscode_2015)))
]

df_filtered
```

```python
# engineer new columns
df_engineered = df_filtered[["year", "unitid", "sportscode"]].copy()
df_engineered["perc_partic_men"] = df_filtered["sum_partic_men"] / (df_filtered["sum_partic_men"] + df_filtered["sum_partic_women"])
df_engineered["perc_exp_men"] = df_filtered["exp_men"] / df_filtered["total_exp_menwomen"]
df_engineered
```

```python
# make combined df
df_final = df_engineered[df_engineered.year == 2015].copy()
df_final = df_final[["unitid", "sportscode"]]
df_final["pp_men_2015"] = df_engineered[df_engineered["year"] == 2015]["perc_partic_men"]
df_final["pp_men_2019"] = df_engineered[df_engineered["year"] == 2015][["unitid", "sportscode"]].apply(lambda x: df_engineered[
    (df_engineered["year"] == 2019)
    &
    (df_engineered["unitid"] == x["unitid"])
    &
    (df_engineered["sportscode"] == x["sportscode"])
]["perc_partic_men"], axis=1)
df_final
```
